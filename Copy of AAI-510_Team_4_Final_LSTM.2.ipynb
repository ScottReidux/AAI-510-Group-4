```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Dropout
import numpy as np
import matplotlib.pyplot as plt

# Load the CSV file into a DataFrame
df = pd.read_csv('/content/combined_stocks.csv')

# Convert the 'Date' column to datetime type
df['Date'] = pd.to_datetime(df['Date'])

# Filtering data for the relevant years
df = df[(df['Date'] < '2019-01-01')]

# List of tickers to process
tickers = df['Ticker'].unique()

# Define LSTM model parameters
n_past = 60  # Number of past days to consider for predicting the next day's price

results = {}

for ticker in tickers:
    print(f"Processing {ticker}")
    df_ticker = df[df['Ticker'] == ticker]
    df_ticker.set_index('Date', inplace=True)

    # Prepare data for LSTM
    scaler = MinMaxScaler(feature_range=(0, 1))
    close_prices = df_ticker['Close'].values.reshape(-1, 1)
    scaled_data = scaler.fit_transform(close_prices)

    # Create the training data
    X_train, y_train = [], []
    for i in range(n_past, len(scaled_data)-365):  # Reserve the last year for testing
        X_train.append(scaled_data[i-n_past:i, 0])
        y_train.append(scaled_data[i, 0])
    X_train, y_train = np.array(X_train), np.array(y_train)
    X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))

    # Build the LSTM model
    model = Sequential()
    model.add(LSTM(units=50, return_sequences=True, input_shape=(X_train.shape[1], 1)))
    model.add(Dropout(0.2))
    model.add(LSTM(units=50, return_sequences=False))
    model.add(Dropout(0.2))
    model.add(Dense(1))

    # Compile and fit the model
    model.compile(optimizer='adam', loss='mean_squared_error')
    model.fit(X_train, y_train, epochs=20, batch_size=32)

    # Predicting 2017-2018 values
    test_data = scaled_data[-(n_past+365):]  # Last year data + 60 days
    X_test = []
    for i in range(n_past, len(test_data)):
        X_test.append(test_data[i-n_past:i, 0])
    X_test = np.array(X_test)
    X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
    predicted_stock_price = model.predict(X_test)
    predicted_stock_price = scaler.inverse_transform(predicted_stock_price)

    # Actual closing prices
    actual_prices = df_ticker['Close'].tail(365).values






    # Calculate statistics
    mse = mean_squared_error(actual_prices, predicted_stock_price)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(actual_prices, predicted_stock_price)
    mape = np.mean(np.abs((actual_prices - predicted_stock_price) / actual_prices)) * 100
    r2 = r2_score(actual_prices, predicted_stock_price)

    # Print statistics
    print(f"Metrics for {ticker}: MSE: {mse}, RMSE: {rmse}, MAE: {mae}, MAPE: {mape}%, R^2: {r2}")








    # Plotting the results
    plt.figure(figsize=(10, 6))
    plt.plot(df_ticker.index[-365:], actual_prices, color='blue', label='Actual Stock Price')
    plt.plot(df_ticker.index[-365:], predicted_stock_price, color='red', label='Predicted Stock Price')
    plt.title(f'Stock Price Prediction for {ticker}')
    plt.xlabel('Date')
    plt.ylabel('Stock Price')
    plt.legend()
    plt.show()

    # Store results
    results[ticker] = predicted_stock_price
```
